```python
import pandas as pd

# 假設 EMS 是您的 DataFrame

# 檢查每個 transgress_control_id 的 fac_address 是否一致
same_address_results = {}
for control_id, group in EMS.groupby('transgress_control_id'):
    # 去除 fac_address 欄位中的所有特殊字元，只保留字母、數字和中文字符
    group['fac_uniformno'] = group['fac_uniformno'].str.replace(r'[^a-zA-Z0-9\u4e00-\u9fa5]', '', regex=True)
    
    # 檢查當前 group 中 fac_address 欄位的唯一值是否為 1（表示都一樣）
    if group['fac_uniformno'].nunique() == 1:
        same_address_results[control_id] = True  # 如果都相同
    else:
        same_address_results[control_id] = False  # 如果不相同
        # 輸出具有不同 fac_address 的 transgress_control_id 和其對應的 fac_address
        print(f"transgress_control_id = {control_id} 的 fac_address 有不同的值：")
        print(group[['fac_uniformno']].drop_duplicates(), "\n")

# 檢查所有不同的 transgress_control_id 的 fac_address 是否互不相同
different_address_results = {}
unique_fac_addresses = EMS.groupby('transgress_control_id')['fac_uniformno'].apply(
    lambda x: x.str.replace(r'[^a-zA-Z0-9\u4e00-\u9fa5]', '', regex=True).unique())

for control_id, fac_addresses in unique_fac_addresses.items():
    if len(fac_addresses) == 1:
        different_address_results[control_id] = True  # 如果都是唯一的（只有一種）
    else:
        different_address_results[control_id] = False  # 如果有多個不同的值
        # 輸出在同一個 transgress_control_id 中有多個不同的 fac_address 資料
        print(f"transgress_control_id = {control_id} 有多個不同的 fac_address：")
        print(fac_addresses, "\n")

# 檢查兩個結果是否相同
same_result = all(value == True for value in same_address_results.values())
different_result = all(value == True for value in different_address_results.values())

# 輸出結果
if same_result and different_result:
    print("所有結果都相同且所有結果都不同")
elif same_result:
    print("所有結果都相同")
elif different_result:
    print("所有結果都不同")
else:
    print("結果有部分相同，部分不同")
```

```python
import pandas as pd

# 自定義函式：從 fac_uniformno 抽取後綴
def extract_suffix(fac_uniformno):
    if pd.isna(fac_uniformno):
        return 'nan'
    return ''.join([c for c in str(fac_uniformno) if c.isdigit() or c.lower() == 'n'])

# 主分析函式
def check_redundancy(df, key_col='fac_uniformno', exclude_cols=['fac_uniformno', 'fac_name', 'document_no', 'transgress_name']):
    df = df.copy()
    df['uniform_suffix'] = df[key_col].apply(extract_suffix)

    result_log = []

    for suffix, group in df.groupby('uniform_suffix'):
        subgroup_dict = {}
        for prefix_value, subdf in group.groupby(key_col):
            subgroup_dict[prefix_value] = subdf.reset_index(drop=True)

        keys = list(subgroup_dict.keys())

        # 情境 A：前綴不同但後綴相同
        if len(keys) > 1:
            base = subgroup_dict[keys[0]]
            for other_key in keys[1:]:
                other = subgroup_dict[other_key]
                if len(base) == len(other):
                    base_cmp = base.drop(columns=exclude_cols, errors='ignore').reset_index(drop=True)
                    other_cmp = other.drop(columns=exclude_cols, errors='ignore').reset_index(drop=True)
                    identical = base_cmp.equals(other_cmp)
                    result_log.append({
                        'suffix': suffix,
                        'type': 'prefix_diff_same_len',
                        'keys': (keys[0], other_key),
                        'equal': identical,
                        'count': len(base)
                    })

        # 情境 B：同前綴但筆數為 2 倍（懷疑重複儲存）
        for key, group_df in subgroup_dict.items():
            count = len(group_df)
            for other_key, other_df in subgroup_dict.items():
                if key == other_key:
                    continue
                if other_key.startswith(key) and len(other_df) == count * 2:
                    first_half = other_df.iloc[:count].drop(columns=exclude_cols, errors='ignore').reset_index(drop=True)
                    second_half = other_df.iloc[count:].drop(columns=exclude_cols, errors='ignore').reset_index(drop=True)
                    base_cmp = group_df.drop(columns=exclude_cols, errors='ignore').reset_index(drop=True)
                    is_redundant = base_cmp.equals(first_half) and base_cmp.equals(second_half)
                    result_log.append({
                        'suffix': suffix,
                        'type': 'prefix_same_double_count',
                        'keys': (key, other_key),
                        'equal': is_redundant,
                        'base_count': count,
                        'double_count': len(other_df)
                    })

    return pd.DataFrame(result_log)
```
